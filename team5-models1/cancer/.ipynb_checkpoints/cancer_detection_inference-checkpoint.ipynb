```python
import pandas as pd
import joblib
```

# Read Data

```python
csv_test = pd.read_csv('./data.csv')
```

```python
x_test = csv_test.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1)
```

```python
x_test
```

# Load Model

```python
svc_model = joblib.load('cancer_detection_model_svm.pkl')
```

# Run Model

## single data

```python
predict = svc_model.predict(x_test.iloc[11:12])
```

```python
predict
```

## batch data

```python
predict = svc_model.predict(x_test.iloc[1:101])
```

```python
predict
```

```python
from minio import Minio
```

```python
minioClient = Minio('34.145.3.171:32001',
                     access_key='minio',
                     secret_key='minio123',
                     secure=False)
```

```python
with open('cancer_detection_model_svm.pkl', 'rb') as file_data:
    file_stat = os.stat('cancer_detection_model_svm.pkl')
    minioClient.put_object('model', 'cancer/1', file_data,
                    file_stat.st_size)
```

```python
objects = minioClient.list_objects('model', recursive=True)
```

```python
for obj in objects:
    print(obj.bucket_name, obj.object_name.encode('utf-8'), obj.last_modified,
          obj.etag, obj.size, obj.content_type)
```
